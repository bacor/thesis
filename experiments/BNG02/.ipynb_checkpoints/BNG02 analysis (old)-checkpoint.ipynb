```python
# Imports
%pylab inline
import sys
sys.path.append('../../naming-games/')
from helpers import *
import pickle
import json
```

```python
# Load all data
batch = range(1, 4)
results = {}
for B in Bs:
    fn = 'HUR01/HUR01-B-{}.pickle'.format(B)
    bases, successes = pickle.load(open(fn, 'rb'))
    bases = array(bases)
    successes = array(successes)
    results[B] = {
        'bases': array(bases),
        'successes': array(successes)
    }

    results[B]['base_counts'] = bases.sum(axis=2)
    results[B]['total_base_count'] = bases.sum(axis=3).sum(axis=2)
    results[B]['unique_base_count'] = (bases.sum(axis=2) > 0).sum(axis=2)
```

```python
jsd_thetas, jsd_alpha_thetas, jsd_alpha_mean_theta, xs, counts = pickle.load(open('results/test.pickle', 'rb'))
jsd_thetas = array(jsd_thetas).mean(axis=0) 
jsd_alpha_thetas = array(jsd_alpha_thetas).mean(axis=0)
jsd_alpha_mean_theta = array(jsd_alpha_mean_theta).mean(axis=0)
xs = xs[0]
```

```python
figure(figsize=cm2inch(6,4))
pyplot.loglog(xs, jsd_alpha_mean_theta, label=r'$\mathsf{JSD}(\alpha, \overline{\phi})$')
pyplot.loglog(xs, jsd_alpha_thetas, label=r'$\mathsf{JSD}(\alpha, \phi_1, \dots, \phi_N)$')
pyplot.loglog(xs, jsd_thetas, label=r'$\mathsf{JSD}(\phi_1, \dots, \phi_N)$')
legend(loc=8)
xlabel('time')
ylabel('divergence')
sns.despine()
```

```python
# Plot
data = pickle.load(open('HUR01-plotdata.pickle', 'rb'))
figure(figsize=cm2inch(12,3))
xs = arange(0, setup['T'], setup['res'])
dashes = {30: (1,1), 20: (4,2,1,2), 10: (4,2.5), 1:()}

def plot_results(name):
    for B in [1, 10, 20, 30]:
        mean = data[B][name]['mean']
        std = data[B][name]['std']

        plot(xs, mean, 'k', dashes=dashes[B], label='B = {}'.format(B))
        fill_between(xs, mean-.5*std, mean+.5*std, 
                         alpha=.4, facecolor='#ed6751', edgecolor='#ffffff')
ax1 = subplot(121)
plot_results('total_base_count')
title('Total base count')
xlabel('time')

ax2 = subplot(122)
plot_results('unique_base_count')
title('Unique base count')
xlabel('time')
legend()

# Format axes
restyle_title(ax1, ax2)
for ax in [ax1, ax2]:
    ax.locator_params(nbins=5, axis='y')
    ax.locator_params(nbins=5, axis='x')
    ax.yaxis.grid()
    ax.set_xlim(0, 3500)
sns.despine()

# Remove all margins, so the plots have exactly the right dimensions
subplots_adjust(left=0, bottom=0, right=1, top=1)
savefig('HUR01b-results.pdf', bbox_inches='tight')
```
